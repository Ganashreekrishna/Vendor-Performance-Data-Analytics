```python
vendor_sales_summary = pd.read_sql_query(
    """
    WITH FreightSummary AS (
        SELECT
            VendorNumber,
            SUM(Freight) AS FreightCost
        FROM vendor_invoice
        GROUP BY VendorNumber
    ),

    PurchaseSummary AS (
        SELECT
            p.VendorNumber,
            p.VendorName,
            p.Brand,
            p.Description,
            p.PurchasePrice,
            pp.Price AS ActualPrice,
            pp.Volume,
            SUM(p.Quantity) AS TotalPurchaseQuantity,
            SUM(p.Dollars) AS TotalPurchaseDollars
        FROM purchases p
        JOIN purchase_prices pp
            ON p.Brand = pp.Brand
        WHERE p.PurchasePrice > 0
        GROUP BY
            p.VendorNumber,
            p.VendorName,
            p.Brand,
            p.Description,
            p.PurchasePrice,
            pp.Price,
            pp.Volume
    ),

    SalesSummary AS (
        SELECT
            VendorNo AS VendorNumber,
            Brand,
            SUM(SalesQuantity) AS TotalSalesQuantity,
            SUM(SalesDollars) AS TotalSalesDollars,
            SUM(SalesPrice) AS TotalSalesPrice,
            SUM(ExciseTax) AS TotalExciseTax
        FROM sales
        GROUP BY VendorNo, Brand
    )

    SELECT
        ps.VendorNumber,
        ps.VendorName,
        ps.Brand,
        ps.Description,
        ps.PurchasePrice,
        ps.ActualPrice,
        ps.Volume,
        ps.TotalPurchaseQuantity,
        ps.TotalPurchaseDollars,
        ss.TotalSalesQuantity,
        ss.TotalSalesDollars,
        ss.TotalSalesPrice,
        ss.TotalExciseTax,
        fs.FreightCost
    FROM PurchaseSummary ps
    LEFT JOIN SalesSummary ss
        ON ps.VendorNumber = ss.VendorNumber
        AND ps.Brand = ss.Brand
    LEFT JOIN FreightSummary fs
        ON ps.VendorNumber = fs.VendorNumber
    ORDER BY ps.TotalPurchaseDollars DESC
    """,
    conn
)

vendor_sales_summary
```

```python
# Convert Volume column to float for numerical analysis
vendor_sales_summary['Volume'] = vendor_sales_summary['Volume'].astype('float64')
```

```python
# Replace all missing (NaN) values with 0
vendor_sales_summary.fillna(0, inplace=True)
```

```python
# Remove leading and trailing spaces from VendorName
vendor_sales_summary['VendorName'] = vendor_sales_summary['VendorName'].str.strip()
```

```python
# Gross Profit = Total Sales Revenue − Total Purchase Cost
# Shows absolute profit made per vendor–brand
vendor_sales_summary['GrossProfit'] = (
    vendor_sales_summary['TotalSalesDollars']
    - vendor_sales_summary['TotalPurchaseDollars']
)
```

```python
# Profit Margin (%) = (Gross Profit / Total Sales Revenue) × 100
# Indicates profitability efficiency
vendor_sales_summary['ProfitMargin'] = (
    vendor_sales_summary['GrossProfit']
    / vendor_sales_summary['TotalSalesDollars']
) * 100
```

```python
# Stock Turnover = Total Sales Quantity / Total Purchase Quantity
# Measures how efficiently inventory is sold
vendor_sales_summary['StockTurnover'] = (
    vendor_sales_summary['TotalSalesQuantity']
    / vendor_sales_summary['TotalPurchaseQuantity']
)
```

```python
# Sales-to-Purchase Ratio = Total Sales Revenue / Total Purchase Cost
# Measures return per unit of purchasing spend
vendor_sales_summary['SalesToPurchaseRatio'] = (
    vendor_sales_summary['TotalSalesDollars']
    / vendor_sales_summary['TotalPurchaseDollars']
)
```

```python
%%writefile get_vendor_summary.py
import sqlite3
import pandas as pd

def get_vendor_summary(db_path):
    """
    Fetch vendor performance summary from SQLite database
    """
    conn = sqlite3.connect(db_path)

    query = """
    SELECT
        VendorNumber,
        VendorName,
        Brand,
        TotalPurchaseDollars,
        TotalSalesDollars,
        GrossProfit,
        ProfitMargin,
        StockTurnover,
        SalesToPurchaseRatio
    FROM vendor_sales_summary
    ORDER BY GrossProfit DESC
    """

    df = pd.read_sql_query(query, conn)
    conn.close()
    return df


if __name__ == "__main__":
    db_path = "/content/drive/MyDrive/inventory.db"
    result = get_vendor_summary(db_path)
    print(result.head(10))
```

```python
%%writefile /content/drive/MyDrive/get_vendor_summary.py

import os
import sqlite3
import pandas as pd
import logging
from ingestion_db import ingest_db

# ------------------ LOGGING ------------------
LOG_DIR = "/content/drive/MyDrive/logs"
os.makedirs(LOG_DIR, exist_ok=True)

logging.basicConfig(
    filename=os.path.join(LOG_DIR, "get_vendor_summary.log"),
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s",
    filemode="a"
)

# ------------------ FUNCTIONS ------------------
def create_vendor_summary(conn):
    """
    Creates a consolidated vendor-level summary by combining
    purchase, sales, and freight data.
    """

    query = """
    WITH FreightSummary AS (
        SELECT
            VendorNumber,
            SUM(Freight) AS FreightCost
        FROM vendor_invoice
        GROUP BY VendorNumber
    ),

    PurchaseSummary AS (
        SELECT
            p.VendorNumber,
            p.VendorName,
            p.Brand,
            p.Description,
            p.PurchasePrice,
            pp.Price AS ActualPrice,
            pp.Volume,
            SUM(p.Quantity) AS TotalPurchaseQuantity,
            SUM(p.Dollars) AS TotalPurchaseDollars
        FROM purchases p
        JOIN purchase_prices pp
            ON p.Brand = pp.Brand
        WHERE p.PurchasePrice > 0
        GROUP BY
            p.VendorNumber,
            p.VendorName,
            p.Brand,
            p.Description,
            p.PurchasePrice,
            pp.Price,
            pp.Volume
    ),

    SalesSummary AS (
        SELECT
            VendorNo,
            Brand,
            SUM(SalesQuantity) AS TotalSalesQuantity,
            SUM(SalesDollars) AS TotalSalesDollars,
            SUM(SalesPrice) AS TotalSalesPrice,
            SUM(ExciseTax) AS TotalExciseTax
        FROM sales
        GROUP BY VendorNo, Brand
    )

    SELECT
        ps.VendorNumber,
        ps.VendorName,
        ps.Brand,
        ps.Description,
        ps.PurchasePrice,
        ps.ActualPrice,
        ps.Volume,
        ps.TotalPurchaseQuantity,
        ps.TotalPurchaseDollars,
        ss.TotalSalesQuantity,
        ss.TotalSalesDollars,
        ss.TotalSalesPrice,
        ss.TotalExciseTax,
        fs.FreightCost
    FROM PurchaseSummary ps
    LEFT JOIN SalesSummary ss
        ON ps.VendorNumber = ss.VendorNo
        AND ps.Brand = ss.Brand
    LEFT JOIN FreightSummary fs
        ON ps.VendorNumber = fs.VendorNumber
    ORDER BY ps.TotalPurchaseDollars DESC
    """

    return pd.read_sql_query(query, conn)


def clean_data(df):
    """
    Cleans data and creates analytical metrics.
    """

    df.fillna(0, inplace=True)

    df["GrossProfit"] = df["TotalSalesDollars"] - df["TotalPurchaseDollars"]
    df["ProfitMargin"] = (df["GrossProfit"] / df["TotalSalesDollars"]) * 100
    df["StockTurnover"] = df["TotalSalesQuantity"] / df["TotalPurchaseQuantity"]
    df["SalesToPurchaseRatio"] = df["TotalSalesDollars"] / df["TotalPurchaseDollars"]

    return df


# ------------------ MAIN ------------------
if __name__ == "__main__":

    logging.info("Starting vendor summary generation")

    conn = sqlite3.connect("/content/drive/MyDrive/inventory.db")

    summary_df = create_vendor_summary(conn)
    logging.info("Vendor summary created")

    final_df = clean_data(summary_df)
    logging.info("Data cleaned and metrics calculated")

    ingest_db(final_df, "vendor_sales_summary", conn)
    logging.info("Vendor summary ingested into database")

    conn.close()
    logging.info("Process completed successfully")
```
