```python
%%writefile /content/drive/MyDrive/ingestion_db.py

import pandas as pd

def ingest_db(df, table_name, conn):
    """
    Ingests a pandas DataFrame into an SQLite database table.
    If the table exists, it will be replaced.
    """
    df.to_sql(
        table_name,
        conn,
        if_exists="replace",
        index=False
    )
```
